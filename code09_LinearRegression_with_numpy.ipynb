# Linear Regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import pandas as pd
```

```python
area = [1.3,4.3,2.3,2.8,3.5,3.9,4.9,5.2,6.4,5.8,5.1,6.7,6.1,7.9,7.2,8.0,8.1,2.7,3.7,3.9]
price= [190,390,270,250,290,450,520,580,550,510,590,750,650,680,650,850,840,310,350,380]
len(area),len(price)
```

```python
x = np.array(area)
y = np.array(price)

LR = 0.0015

m = 10
c = 5

for i in range(10):   
    # get predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y-yhat))
    
    #Update m and c
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    print(error)
    
    plt.scatter(x,y,c='red')
    plt.plot(x,yhat,c='green')
    plt.show()
```

```python
x = np.array(area)
y = np.array(price)

LR = 0.0015

m = 100
c = 500

for i in range(1000):
    
    # get predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y-yhat))
    
    #Update m and c
    
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(error)
```

```python
print(m,c)
```

# Linear Regression with Scikit-Learn

```python
area = [1.3,4.3,2.3,2.8,3.5,3.9,4.9,5.2,6.4,5.8,5.1,6.7,6.1,7.9,7.2,8.0,8.1,2.7,3.7,3.9]
price= [190,390,270,250,290,450,520,580,550,510,590,750,650,680,650,850,840,310,350,380]
len(area),len(price)
```

```python
x = np.array(area)
x.shape
```

```python
x = np.array(area).reshape(-1,1)
x.shape
```

```python
x = np.array(x).reshape(-1,1)
y = np.array(y)


from sklearn.linear_model import LinearRegression
model = LinearRegression()


# train the model with trianing data
model.fit(x,y)
```

```python
print(model.coef_,model.intercept_)
```
