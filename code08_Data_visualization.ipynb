# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df= pd.read_csv("datasets-1/profit_forecast_reported.csv")
df.head()
```

```python
#line plot
plt.plot(df.Day,df.Forecasted_Profit)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.Day,df.Forecasted_Profit)
plt.show()
```

```python
sns.set_theme()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.Day,df.Forecasted_Profit)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.Day,df.Forecasted_Profit,c='red',label='Forecasted Profit')
plt.xlabel("Days")
plt.ylabel("Profit Data")
plt.title("Forecasted and Reported Profit")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.Day,df.Forecasted_Profit,c='green',label='Forecasted Profit')
plt.plot(df.Day,df.Reported_Profit_Source1,c='blue',label='Reported Profit Source 1')
plt.plot(df.Day,df.Reported_Profit_Source2,c='red',label='Reported Profit Source 2')
plt.xlabel("Days")
plt.ylabel("Profit Data")
plt.title("Forecasted and Reported Profit")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.Day,df.Forecasted_Profit,c='green',label='Forecasted Profit')
plt.scatter(df.Day,df.Reported_Profit_Source1,c='blue',label='Reported Profit Source 1')
plt.scatter(df.Day,df.Reported_Profit_Source2,c='red',label='Reported Profit Source 2')
plt.xlabel("Days")
plt.ylabel("Profit Data")
plt.title("Forecasted and Reported Profit")
plt.legend()
plt.show()
```
